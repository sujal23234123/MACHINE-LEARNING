# FILLING MISSING VALUES


```python
dataset.isnull().sum()
```

```python
dataset.head()
```

# FOR INTEGER VALUE FILL WITH MEDIAN
# FOR CATEGORIE DATA FILL WITH MODE

```python

warnings.filterwarnings("ignore")

#AGE
dataset.Age.fillna(dataset.Age.median(),inplace=True)

#Type of contact
dataset.TypeofContact.fillna(dataset.TypeofContact.mode()[0],inplace=True)

#Duration of pitch
dataset.DurationOfPitch.fillna(dataset.DurationOfPitch.mode()[0],inplace=True)

#numberoffollowups
dataset.NumberOfFollowups.fillna(dataset.NumberOfFollowups.median(),inplace=True)

#PreferredPropertyStar	
dataset.PreferredPropertyStar.fillna(dataset.PreferredPropertyStar.median(),inplace=True)	

#NumberOfTrips
dataset.NumberOfTrips.fillna(dataset.NumberOfTrips.median(),inplace=True)

#NumberOfChildrenVisiting
dataset.NumberOfChildrenVisiting.fillna(dataset.NumberOfChildrenVisiting.median(),inplace=True)

#MonthlyIncome
dataset.MonthlyIncome.fillna(dataset.MonthlyIncome.median(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

# COMBINE NO OF PERSON AND CHILDREN THAT ARE VISITING

```python
dataset["Total visiting"]=dataset["NumberOfPersonVisiting"]+dataset["NumberOfChildrenVisiting"]
```

```python
dataset.drop(columns=["NumberOfPersonVisiting","NumberOfChildrenVisiting"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset.drop("CustomerID",axis=1,inplace=True)
```

# Train test split

```python
X=dataset.drop(["ProdTaken"],axis=1)
Y=dataset["ProdTaken"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=40)
```

# Encoding

```python
categories=X.select_dtypes(include="object").columns
numeric=X.select_dtypes(exclude="object").columns

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

cat_transform=OneHotEncoder(handle_unknown='ignore', )
num_transform=StandardScaler()

preprocessing=ColumnTransformer(
      [
          ("OneHotEncoder",cat_transform,categories),
          ("StandardScaler",num_transform,numeric)
      ]
)
```

```python
x_train=preprocessing.fit_transform(x_train)
```

```python
pd.DataFrame(x_train)
```

```python
x_test=preprocessing.transform(x_test)
```

```python
x_test
```

```python
x_train
```

# RANDOM FOREST

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report,
    r2_score, f1_score, roc_auc_score, precision_score
)

models = {
    
    "Random Forest": RandomForestClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "Gredient boosting":GradientBoostingClassifier()
}

for name, model in models.items():
    model.fit(x_train, y_train)

    train_pred = model.predict(x_train)
    test_pred = model.predict(x_test)

    print(f"\n{name}")
    print("-" * 50)

    print("accuracy score of train data is :", accuracy_score(y_train, train_pred))
    print("r2 score of train data is :", r2_score(y_train, train_pred))
    print("f1 score of train data is :", f1_score(y_train, train_pred))
    print("roc_auc_score of train data is :", roc_auc_score(y_train, train_pred))
    print("the precision score of train data is :", precision_score(y_train, train_pred))
    print("the confusion matrix of train data is :", confusion_matrix(y_train, train_pred))
    print("the classification report of train data is :\n", classification_report(y_train, train_pred))

    print("\n" + "-" * 50)

    print("accuracy score of test data is :", accuracy_score(y_test, test_pred))
    print("r2 score of test data is :", r2_score(y_test, test_pred))
    print("f1 score of test data is :", f1_score(y_test, test_pred))
    print("roc_auc_score of test data is :", roc_auc_score(y_test, test_pred))
    print("the precision score of test data is :", precision_score(y_test, test_pred))
    print("the confusion matrix of test data is :", confusion_matrix(y_test, test_pred))
    print("the classification report of test data is :\n",classification_report(y_test,test_pred))
    
```

# HYPER PARAMATER TUNING

```python
rf_params={"max_depth" :[5,8,15,None,10],
           "max_features":[5,7,"auto",15],
           "min_samples_split":[2,6,10,15],
           "n_estimators":[100,200,500,1000]
           }

gredient_params={"loss":["log_loss", "exponential"],
                 "n_estimators":[100,200,500,1000],
                 "criterion":["friedman_mse", "squared_error"]
}
```

```python
rf_params
```

```python
gredient_params
```

```python
randomcv_models=[
            ("RF",RandomForestClassifier(),rf_params),
            ("GB",GradientBoostingClassifier(),gredient_params)
            
]
```

```python
randomcv_models
```

```python
from sklearn.model_selection import RandomizedSearchCV

model_params={}

for name,model,rf_params in randomcv_models:
    random=RandomizedSearchCV(estimator=model,param_distributions=rf_params,n_iter=100,verbose=2,cv=3,n_jobs=-1)

    random.fit(x_train,y_train)
    model_params[name]=random.best_params_

for model_name  in model_params:
    print(f"----------------Best params for {model_name}----------------------")
    print(model_params[model_name])
```
