```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# CREATING DATASET

```python
from sklearn.datasets import make_classification

X,Y=make_classification(n_samples=1000,n_classes=2,n_features=2,random_state=42,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
pd.DataFrame(X)[0]
pd.DataFrame(X)[1]
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

# MODEL TRAINING

```python
from sklearn.svm import SVC

svc=SVC(kernel="linear")

svc.fit(x_train,y_train)
pridicted=svc.predict(x_test)
```

```python
pridicted
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm=confusion_matrix(y_test,pridicted)
print("The confusion matrix is :",cm)

cr=classification_report(y_test,pridicted)
print("The classifiction report is  :",cr)

ac=accuracy_score(y_test,pridicted)
print(f"The accuracy score is :{ac*100}%")
```

# IF WE HAVE NON LINEAR DATASET THEN USE RBF

```python
from sklearn.datasets import make_classification

X1,Y1=make_classification(n_classes=2,n_features=2,n_samples=1000,n_clusters_per_class=2,n_redundant=0,random_state=10)
```

```python
X1
```

```python
Y1
```

```python
from sklearn.model_selection import train_test_split

x_train1,x_test1,y_train1,y_test1=train_test_split(X1,Y1,test_size=0.30,random_state=20)
```

```python
from sklearn.svm import SVC

rbf=SVC(kernel="rbf")

rbf.fit(x_train1,y_train1)

pridicted1=rbf.predict(x_test1)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm=confusion_matrix(y_test1,pridicted1)
print("The confusion matrix is :",cm)

cr=classification_report(y_test1,pridicted1)
print("The classifiction report is  :",cr)

ac=accuracy_score(y_test1,pridicted1)
print(f"The accuracy score is :{ac*100}%")
```

# FOR POLYNOMIAL DATA

```python
from sklearn.datasets import make_classification

X,Y=make_classification(n_classes=2,n_features=2,n_samples=1000,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split

x_train2,x_test2,y_train2,y_test2=train_test_split(X,Y,test_size=0.30,random_state=40)

```

# MODEL TRAINING    

```python
from sklearn.svm import SVC

poly=SVC(kernel="poly")

poly.fit(x_train2,y_train2)

predict3=poly.predict(x_test2)

```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix


ac=accuracy_score(y_test2,predict3)
print(ac)

cr=classification_report(y_test2,predict3)
print(cr)

cm=confusion_matrix(y_test2,predict3)
print(cm)
```

# HYPERPARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV

param_grid={ "C" :[100,10,1.0,0.1,0.01],
            "gamma" :[100,10,1,0,0.1,0.01],
            "kernel" :["rbf"]
}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,verbose=3,refit=True,cv=5)
```

```python
grid.fit(x_train2,y_train2)
```

```python
predict4=grid.predict(x_test2)

```

```python
grid.best_params_
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix


ac=accuracy_score(y_test2,predict4)
print(ac*100)

cr=classification_report(y_test2,predict4)
print(cr)

cm=confusion_matrix(y_test2,predict4)
```
