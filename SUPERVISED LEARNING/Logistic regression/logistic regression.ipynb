```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# CREATING A DATASET USING SKLEARN

```python
from sklearn.datasets import make_classification

X,Y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=40)
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=42)

x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# MODEL TRAINING


```python
from sklearn.linear_model import LogisticRegression

logistic=LogisticRegression()

logistic.fit(x_train,y_train)
```

```python
pridicted=logistic.predict(x_test)

pridicted
```

# PERFORMANCE METRICS

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

ac=accuracy_score(y_test,pridicted)
print(f"The accuracy score is :{ac*100}%")

cm=confusion_matrix(y_test,pridicted)
print("The confusion matrix is : ",cm)

cr=classification_report(y_test,pridicted)
print("the classfication report is : ",cr)
```

# HYPERPARAMETER TUNING AND CROSS VALIDATION

```python
model=LogisticRegression()
penalty=('l1', 'l2', 'elasticnet')
c_values=(100,10,1,0.1,0.01)
solver=('lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga')
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

cv=StratifiedKFold()

grid=GridSearchCV(estimator=model,param_grid=params,scoring="accuracy",cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_*100
```

```python
gridpridicted=grid.predict(x_test)
```

```python
ac=accuracy_score(y_test,gridpridicted)
print(f"The accuracy score is :{ac*100}%")

cm=confusion_matrix(y_test,gridpridicted)
print("The confusion matrix is : ",cm)

cr=classification_report(y_test,gridpridicted)
print("the classfication report is : ",cr)
```

# RANDOMIZED SEARCHCV

```python
model=LogisticRegression()

from sklearn.model_selection import RandomizedSearchCV

randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,scoring="accuracy",cv=5)
```

```python
randomcv.fit(x_train,y_train)
```

```python
randomcv.best_score_*100
```

```python
randomcv.best_params_
```

```python
pridictedrandom=randomcv.predict(x_test)
```

```python
ac=accuracy_score(y_test,pridictedrandom)
print(f"The accuracy score is :{ac*100}%")

cm=confusion_matrix(y_test,pridictedrandom)
print("The confusion matrix is : ",cm)

cr=classification_report(y_test,pridictedrandom)
print("the classfication report is : ",cr)
```

# LOGISTIC REGRESSION OVR(ONE OVER REST)

```python
X,Y=make_classification(n_samples=1000,n_classes=3,n_informative=3,n_features=10,random_state=42)
```

```python
X
```

```python
Y
```

# TRAIN TEST SPLIT

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
logistic=LogisticRegression(multi_class="ovr")

logistic.fit(x_train,y_train)
```

```python
pridictedovr=logistic.predict(x_test)
```

```python
ac=accuracy_score(y_test,pridictedovr)
print(f"The accuracy score is :{ac*100}%")

cm=confusion_matrix(y_test,pridictedovr)
print("The confusion matrix is : ",cm)

cr=classification_report(y_test,pridictedovr)
print("the classfication report is : ",cr)
```
