# FILLING MISSING VALUES


```python
dataset.isnull().sum()
```

```python
dataset.head()
```

# FOR INTEGER VALUE FILL WITH MEDIAN
# FOR CATEGORIE DATA FILL WITH MODE

```python

warnings.filterwarnings("ignore")

#AGE
dataset.Age.fillna(dataset.Age.median(),inplace=True)

#Type of contact
dataset.TypeofContact.fillna(dataset.TypeofContact.mode()[0],inplace=True)

#Duration of pitch
dataset.DurationOfPitch.fillna(dataset.DurationOfPitch.mode()[0],inplace=True)

#numberoffollowups
dataset.NumberOfFollowups.fillna(dataset.NumberOfFollowups.median(),inplace=True)

#PreferredPropertyStar	
dataset.PreferredPropertyStar.fillna(dataset.PreferredPropertyStar.median(),inplace=True)	

#NumberOfTrips
dataset.NumberOfTrips.fillna(dataset.NumberOfTrips.median(),inplace=True)

#NumberOfChildrenVisiting
dataset.NumberOfChildrenVisiting.fillna(dataset.NumberOfChildrenVisiting.median(),inplace=True)

#MonthlyIncome
dataset.MonthlyIncome.fillna(dataset.MonthlyIncome.median(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

# COMBINE NO OF PERSON AND CHILDREN THAT ARE VISITING

```python
dataset["Total visiting"]=dataset["NumberOfPersonVisiting"]+dataset["NumberOfChildrenVisiting"]
```

```python
dataset.drop(columns=["NumberOfPersonVisiting","NumberOfChildrenVisiting"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset.drop("CustomerID",axis=1,inplace=True)
```

# Train test split

```python
X=dataset.drop(["ProdTaken"],axis=1)
Y=dataset["ProdTaken"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=40)
```

# Encoding

```python
categories=X.select_dtypes(include="object").columns
numeric=X.select_dtypes(exclude="object").columns

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

cat_transform=OneHotEncoder(handle_unknown='ignore', )
num_transform=StandardScaler()

preprocessing=ColumnTransformer(
      [
          ("OneHotEncoder",cat_transform,categories),
          ("StandardScaler",num_transform,numeric)
      ]
)
```

```python
x_train=preprocessing.fit_transform(x_train)
```

```python
pd.DataFrame(x_train)
```

```python
x_test=preprocessing.transform(x_test)
```

```python
x_test
```

```python
x_train
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,r2_score,f1_score,roc_auc_score,precision_score


models={
       "Decision Tree":DecisionTreeClassifier(),
       "Randomforest":RandomForestClassifier()
}

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)


train_pred=model.predict(x_train)
test_pred=model.predict(x_test)

model_train_accuracy_score=accuracy_score(y_train,train_pred)
model_train_confusion_matrix=confusion_matrix(y_train,train_pred)
model_train_classification_report=classification_report(y_train,train_pred)
model_train_r2_score=r2_score(y_train,train_pred)
model_train_f1_score=f1_score(y_train,train_pred)
model_train_roc_auc_score=roc_auc_score(y_train,train_pred)
model_train_precision_score=precision_score(y_train,train_pred)

model_test_accuracy_score=accuracy_score(y_test,test_pred)
model_test_confusion_matrix=confusion_matrix(y_test,test_pred)
model_test_classification_report=classification_report(y_test,test_pred)
model_test_r2_score=r2_score(y_test,test_pred)
model_test_f1_score=f1_score(y_test,test_pred)
model_test_roc_auc_score=roc_auc_score(y_test,test_pred)
model_test_precision_score=precision_score(y_test,test_pred)


print(list(models.keys())[i])
print("-------------------------------------")

print("accuracy score of train data is :",model_train_accuracy_score)
print("r2 score of train data is :",model_train_r2_score)
print("f1 score of train data is :",model_train_r2_score)
print("roc_auc_score of train data is :",model_train_roc_auc_score)
print("the precision score of train data is :",model_train_precision_score)
print("the confusion matrix of train data is :",model_train_confusion_matrix)
print("the classification report of train data is :",model_train_classification_report)

print("-----------------------------------------------")

print("accuracy score of test data is :",model_test_accuracy_score)
print("r2 score of test data is :",model_test_r2_score)
print("f1 score of test data is :",model_test_r2_score)
print("roc_auc_score oftest data is :",model_test_roc_auc_score)
print("the precision score oftest data is :",model_test_precision_score)
print("the confusion matrix of test data is :",model_test_confusion_matrix)
print("the classification report of test data is :",model_test_classification_report)
```
